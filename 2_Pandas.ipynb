#_Índices Multiníveis:_

```python
outside = 'G1 G1 G1 G2 G2 G2'.split()
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside, inside)) #junta as duas listas criando uma dupla, logo criamos uma lista de duplas
```

```python
hier_index
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
hier_index
```

```python
df = pd.DataFrame(randn(6,2), index = hier_index, columns = ['A', 'B'])
df
```

#_Tratando dados ausentes:_

```python
df= pd.DataFrame({'A': [1,2,np.nan],
                  "B": [5,np.nan, np.nan],
                  'C' : [1,2,3]})
df
```

```python
#Usando o dropna
df.dropna(axis = 0) #jogo fora as linhas
```

```python
df.dropna(axis = 1) #jogo fora as colunas
```

```python
df.dropna(thresh = 2, axis = 1) #sempre que encontrar 2 ou mais valores ausentes descarto
```

```python
#Usando o fillna (preencher o valores ausentes)
df.fillna("Conteúdo")
```

```python
df['A'].fillna(value = df['A'].mean())
```

```python
#Usando o ffill (preenche com a última informação)
df.ffill()
```

```python
#Usano o bfill (é o inverso, pego a informação debaixo)
#nesse caso não vai funcionar pois há dados nulos que não possuem dados depois deles
```

#_Groupby:_

```python
import pandas as pd
# Cria um DataFrame
data = {'Classe':['Júnior','Júnior','Pleno','Pleno','Sênior','Sênior'],
       'Nome':['Jorge','Carlos','Roberta','Patrícia','Bruno','Vera'],
       'Venda':[200,120,340,124,243,350]}
df = pd.DataFrame(data)
```

```python
df
```

```python
group = df.groupby('Classe')
```

```python
group.sum()
```

```python
group.mean()
```

```python
df.groupby(df['Classe']).mean()
```

```python
df.groupby(df['Classe']).min()
```

```python
df.groupby(df['Classe']).max()
```

```python
df2 = df.copy() #se eu não usar o .copy eu crio um ponteiro que altera o df
df2
```

```python
df2['Venda'] = [150,432,190,230,410,155]
df2
```

```python
df2
```

```python
df3 = pd.concat([df,df2])
df3
```

```python
df3.groupby(['Classe', 'Nome']).sum()
```
