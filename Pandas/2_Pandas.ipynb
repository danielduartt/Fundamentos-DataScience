#_Tratando dados ausentes:_

```python
df= pd.DataFrame({'A': [1,2,np.nan],
                  "B": [5,np.nan, np.nan],
                  'C' : [1,2,3]})
df
```

```python
#Usando o dropna
df.dropna(axis = 0) #jogo fora as linhas
```

```python
df.dropna(axis = 1) #jogo fora as colunas
```

```python
df.dropna(thresh = 2, axis = 1) #sempre que encontrar 2 ou mais valores ausentes descarto
```

```python
#Usando o fillna (preencher o valores ausentes)
df.fillna("Conteúdo")
```

```python
df['A'].fillna(value = df['A'].mean())
```

```python
#Usando o ffill (preenche com a última informação)
df.ffill()
```

#_Groupby:_

```python
import pandas as pd
# Cria um DataFrame
data = {'Classe':['Júnior','Júnior','Pleno','Pleno','Sênior','Sênior'],
       'Nome':['Jorge','Carlos','Roberta','Patrícia','Bruno','Vera'],
       'Venda':[200,120,340,124,243,350]}
df = pd.DataFrame(data)
```

```python
df
```

```python
group = df.groupby('Classe')
```

```python
group.sum()
```

```python
group.mean()
```

```python
df.groupby(df['Classe']).mean()
```

```python
df.groupby(df['Classe']).min()
```

```python
df.groupby(df['Classe']).max()
```

```python
df2 = df.copy() #se eu não usar o .copy eu crio um ponteiro que altera o df
df2
```

```python
df2['Venda'] = [150,432,190,230,410,155]
df2
```

```python
df2
```

```python
df3 = pd.concat([df,df2])
df3
```

```python
df3.groupby(['Classe', 'Nome']).sum()
```

#_Merge, Concat e Join:_

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
df2
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
df3
```

```python
#Concat (dataframes que compartilham as mesmas colunas com indices diferentes)
#juntar em um único dataframe
df4 = pd.concat([df1,df2,df3])
df4
```

```python
df4 = pd.concat([df1,df2,df3], axis = 1) #expansão horizontal
df4
```

```python
esquerda = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

direita = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
#Merge (DF com colunas diferentes e indices iguais )
```

```python
pd.merge(esquerda,direita, on = "key") #mantem a coluna key que é comum entre as duas
```

```python
#Outro caso
esquerda = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

direita = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita, on = ['key1', 'key2'])
```

```python
pd.merge(esquerda, direita, on = ['key1', 'key2'], how = 'outer') #padrão vem inner
```

```python
pd.merge(esquerda, direita, on = ['key1', 'key2'], how = 'right') #padrão vem inner
```

```python
pd.merge(esquerda, direita, on = ['key1', 'key2'], how = 'left') #padrão vem inner
```

```python
#Join
esquerda = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2'])

direita = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
esquerda
```

```python
direita
```

```python
esquerda.join(direita, how = 'outer')
```
